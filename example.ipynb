# Example Usage of the Corpus Pipeline

We have an input directory which contains .nena formatted texts, `example_texts`, and an output directory `example_out`. 
The pipeline class, `CorpusPipeline`, is instanced on a configuration file, which links to a bunch of definitions 
needed by the various parsers. All data up to the static search tools are produced with `.build_corpus`. This method
requires an in-directory (NENA texts) and out-directory, which is populated with documentation.md, tf, and search_tool.

```python
from pipeline.corpus_pipeline import CorpusPipeline

cp = CorpusPipeline('config.json')

cp.build_corpus('example_texts', 'example_out')
```
